### Preprocessing for Modeling

<font size=3>Function that replaces zeros with NaN's.</font><br><br>

<font size = 2>Code Source: https://stackoverflow.com/questions/49575897/cant-replace-0-to-nan-in-python-using-pandas</font>

```python
def zero_to_nan(df):
    for i in range(14,34):
        df_click.iloc[:, i] = df_click.iloc[:, i].replace(0, np.nan)
```

```python
zero_to_nan(df_click)
```

<font size=3>Function that imputes median for NaN's.</font><br><br>

```python
def impute_median(df):
    for i in range(14,34):
        df_click.iloc[:, i] = df_click.iloc[:, i].fillna(df_click.iloc[:, i].median())
        
```

```python
impute_median(df_click)
```

<font size=3>Expand categorical variables to binary classifiers - for ml.</font><br><br>

<font size=2>Code Source: DataCamp - Machine learning with the experts school budgets course.</font>

```python
df_click = pd.get_dummies(df_click, prefix_sep='_', drop_first=True)
df_assess = pd.get_dummies(df_assess, prefix_sep='_', drop_first=True)
```
